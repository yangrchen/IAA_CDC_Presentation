```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import norm
```

```python
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, loc=0, scale=1)

plt.plot(x, y)
plt.axvline(0)
plt.xticks([0], ['True Parameter']);
plt.yticks([]);
plt.title('Sampling Distribution of Parameter Estimate')
```
